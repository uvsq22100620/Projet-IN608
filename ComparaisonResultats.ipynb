```python
def trouve_genes_dans_commuC(num_commu:int, l_commu_cible:int):
    l_l_commu = [l_commu0, l_commu1, l_commu2, l_commu3, l_commu4, l_commu5]
    ind_commu_precedente = [num_commu]

    for i_l_com in range(l_commu_cible, 0, -1):
        l_commu_precedent = l_l_commu[i_l_com-1]
        #print(l_commu_precedent)
        t = [i for i, valeur in enumerate(l_commu_precedent) if valeur in ind_commu_precedente]
        ind_commu_precedente = t
    return ind_commu_precedente

def trouve_genes_dans_commuD(num_commu:int, l_commu_cible:int):
    l_l_commu = [d_commu0, d_commu1, d_commu2, d_commu3, d_commu4, d_commu5]
    ind_commu_precedente = [num_commu]

    for i_l_com in range(l_commu_cible, 0, -1):
        l_commu_precedent = l_l_commu[i_l_com-1]
        #print(l_commu_precedent)
        t = [i for i, valeur in enumerate(l_commu_precedent) if valeur in ind_commu_precedente]
        ind_commu_precedente = t
    return ind_commu_precedente
```

```python
def jaccard_similarity(list1, list2):
    set1 = set(list1)
    set2 = set(list2)
    intersection = len(set1.intersection(set2))
    union = len(set1.union(set2))
    similarity = intersection / union
    return similarity
```
